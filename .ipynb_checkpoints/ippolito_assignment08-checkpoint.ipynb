## Introduction

This data set comes from the UCI Machine Learning Repository and includes evaluations of 115 Swiss engineering students. Data was collected while the students took six exams each. Data was collected about each student's keystrokes, mouse movements, and test times, along with how the student performed on the exam. I chose the data set because it included a good mix of variables, both numerical and categorical, with which to demonstrate various plot types.


```python
# Load libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import os
import re
```

## Load data

```python
# Set column names
colnamesArray = ['session', 'student_id', 'exercise', 'activity', 'start_time', 'end_time',
                'idle_time', 'mouse_wheel', 'mouse_wheel_click', 'mouse_click_left', 
                 'mouse_click_right', 'mouse_movement', 'keystroke']

# Load session data
dir = 'C:/Users/micha/Downloads/EPM_Dataset_2/Data/Processes'
ct = 0
for subdir in os.listdir(dir):

    # Read files in each subdir
    subdir_fullpath = os.path.join(dir, subdir)
    print("Reading", subdir_fullpath)
    for fn in os.listdir(subdir_fullpath):

        # Build path and read data frame
        ct += 1
        fullpath = os.path.join(subdir_fullpath, fn)
        df = pd.read_csv(fullpath, converters={4: str, 5: str})
        df.columns = colnamesArray

        # Append data to df2
        if ct == 1:
            df2 = df.copy()
        else:
            df2 = pd.concat([df2, df])

        # Remove the dataframe from memory
        del df

# Load grade data
print("Reading C:/Users/micha/Downloads/EPM_Dataset_2/Data/final_grades.csv")
dfgr = pd.read_csv("C:/Users/micha/Downloads/EPM_Dataset_2/Data/final_grades.csv")
print()
print("Done reading data")
```

## Exploratory data analysis

```python
# General info
print("Session data shape:", df2.shape)
print("Grades shape:", dfgr.shape)
print()
print("Session data info:")
print(df2.info())
print()
print("Grades info:")
print(dfgr.info())
print()

# Summary stats
print("Session data summary stats:")
print(df2.describe())
print()
print("Grades summary stats:")
print(dfgr.describe())
print()
```

## Data wrangling

```python
# Start times are in this format: 2.10.2014 11:35:32; convert to ISO8601
df2['start_time'] = df2['start_time'].str.replace('\.', '-', regex=True)
df2['end_time'] = df2['end_time'].str.replace('\.', '-', regex=True)
df2['start_time'] = pd.to_datetime(df2['start_time'])
df2['end_time'] = pd.to_datetime(df2['end_time'])
```

```python
# Remove idle time outlier (-425860446967950)
df2.drop(df2[df2.idle_time < 0].index, inplace=True)
```

```python
# Group process data by student_id
df3 = \
    df2.groupby(['student_id']).agg({'idle_time': 'sum'}).merge(
        df2.groupby(['student_id']).agg({'mouse_wheel': 'sum'}), how='outer', on='student_id'
    ).merge(
        df2.groupby(['student_id']).agg({'mouse_wheel_click': 'sum'}), how='outer', on='student_id'
    ).merge(
        df2.groupby(['student_id']).agg({'mouse_click_left': 'sum'}), how='outer', on='student_id'
    ).merge(
        df2.groupby(['student_id']).agg({'mouse_click_right': 'sum'}), how='outer', on='student_id'
    ).merge(
        df2.groupby(['student_id']).agg({'mouse_movement': 'sum'}), how='outer', on='student_id'
    ).merge(
        df2.groupby(['student_id']).agg({'keystroke': 'sum'}), how='outer', on='student_id'
    )
```

```python
# Get mean of each student's test grade (attempt 1 and attempt 2)
df4 = dfgr.groupby(['student_id']).agg({'total': 'mean'})

# Merge grade with process data
df5 = df4.merge(df3, how='outer', on='student_id')

# Peek at data
df5.head()
```

## Visualizations - Part 1 (Matplotlib)

```python
# 1. Change the title and x/y labels

# Histograms of process data grouped by student_id
plt.clf()
(pf, pa) = plt.subplots(2, 4, figsize=(15,5))
plt.suptitle('Process Data Grouped by Student ID')
plt.subplots_adjust(hspace=0.4)
plt.subplot(2, 4, 1)
plt.xlabel('Idle time (ms)')
plt.hist(data=df5, x='idle_time')
plt.subplot(2, 4, 2)
plt.xlabel('Mouse wheel time (ms)')
plt.hist(data=df5, x='mouse_wheel')
plt.subplot(2, 4, 3)
plt.xlabel('Mouse wheel clicks')
plt.hist(data=df5, x='mouse_wheel_click')
plt.subplot(2, 4, 4)
plt.xlabel('Mouse left clicks')
plt.hist(data=df5, x='mouse_click_left')
plt.subplot(2, 4, 5)
plt.xlabel('Mouse right clicks')
plt.hist(data=df5, x='mouse_click_right')
plt.subplot(2, 4, 6)
plt.xlabel('Mouse move distance (px)')
plt.hist(data=df5, x='mouse_movement')
plt.subplot(2, 4, 7)
plt.xlabel('Keystrokes')
plt.hist(data=df5, x='keystroke')
plt.subplot(2, 4, 8)
pa[-1, -1].axis('off')
plt.show()
```

```python
# 2. Change the marker, point colors, and point size
# 3. Place a legend outside of the plot
# 4. Change a legend font size
# 5. Modify Axis Text Ticks/Labels
# 6. Change size of axis Labels
# 7. Add annotations

# Scatter plot of grades by mouse wheel time
plt.clf()
plt.scatter(data=df5, x='mouse_click_left', y='total', color='#00aa00', marker='*', s=50, alpha=.5)  # change markers
plt.scatter(data=df5, x='keystroke', y='total', color='#aa00aa', marker='+', s=20, alpha=.5)         # change markers
plt.title('Left clicks and keystrokes')
plt.ylabel('Mean total grade')
plt.legend(['Left clicks', 'Keystrokes'], loc=(1.04,0), prop={'size': 11})  # place legend outside the plot
plt.xticks(ticks=np.arange(0, 50001, step=10000))                           # change step of x-ticks
plt.yticks(ticks=np.arange(0, 101, step=10))                                # change step of y-ticks
plt.gca().yaxis.label.set_size(11)                                          # set size of y-axis label
plt.annotate(text='outlier', xy=(48000, 85), xytext=(45000, 90))            # annotate the outlier
plt.show()
```

## Visualizations - Part 2 (Seaborn)

```python
# Recreate histograms using seaborn
plt.clf()
(pf, pa) = plt.subplots(2, 4, figsize=(15,5))
plt.suptitle('Process Data Grouped by Student ID')
plt.subplots_adjust(hspace=0.4)
sns.histplot(data=df5, x="idle_time", ax=pa[0,0], bins=10)
pa[0,0].set_xlabel('Idle time (ms)')
sns.histplot(data=df5, x="mouse_wheel", ax=pa[0,1], bins=10)
pa[0,1].set_xlabel('Mouse wheel time (ms)')
sns.histplot(data=df5, x="mouse_wheel_click", ax=pa[0,2], bins=10)
pa[0,2].set_xlabel('Mouse wheel clicks')
sns.histplot(data=df5, x="mouse_click_left", ax=pa[0,3], bins=10)
pa[0,3].set_xlabel('Mouse left clicks')
sns.histplot(data=df5, x="mouse_click_right", ax=pa[1,0], bins=10)
pa[1,0].set_xlabel('Mouse right clicks')
sns.histplot(data=df5, x="mouse_movement", ax=pa[1,1], bins=10)
pa[1,1].set_xlabel('Mouse move distance (px)')
sns.histplot(data=df5, x="keystroke", ax=pa[1,2], bins=10)
pa[1,2].set_xlabel('Keystrokes')
pa[-1, -1].axis('off')
plt.show()
```

```python
# Recreate scatter plot using seaborn;
# seaborn seems to make markers smaller than matplotlib, so double the size here
plt.clf()
sns.scatterplot(data=df5, x='mouse_click_left', y='total', color='#00aa00', marker='*', s=100, alpha=.5)  # change markers
sns.scatterplot(data=df5, x='keystroke', y='total', color='#aa00aa', marker='+', s=40, alpha=.5)          # change markers
plt.title('Left clicks and keystrokes')
plt.ylabel('Mean total grade')
plt.legend(['Left clicks', 'Keystrokes'], loc=(1.04,0), prop={'size': 11})  # place legend outside the plot
plt.xticks(ticks=np.arange(0, 50001, step=10000))                           # change step of x-ticks
plt.yticks(ticks=np.arange(0, 101, step=10))                                # change step of y-ticks
plt.gca().yaxis.label.set_size(11)                                          # set size of y-axis label
plt.annotate(text='outlier', xy=(48000, 85), xytext=(45000, 90))            # annotate the outlier
plt.show()
```

```python
# Show a few more scatter plots to visually look for a correlation between total and the other variables
(pf, pa) = plt.subplots(2, 3, figsize=(20,8))
sns.scatterplot(data=df5, x='idle_time', y='total', ax=pa[0,0])
pa[0,0].set_xlabel('Idle time (ms)')
sns.scatterplot(data=df5, x='mouse_wheel', y='total', ax=pa[0,1])
pa[0,1].set_xlabel('Mouse wheel time (ms)')
sns.scatterplot(data=df5, x='mouse_wheel_click', y='total', ax=pa[0,2])
pa[0,2].set_xlabel('Mouse wheel clicks')
sns.scatterplot(data=df5, x='mouse_click_right', y='total', ax=pa[1,0])
pa[1,0].set_xlabel('Mouse right clicks')
sns.scatterplot(data=df5, x='mouse_movement', y='total', ax=pa[1,1])
pa[1,1].set_xlabel('Mouse move distance (px)')
pa[-1, -1].axis('off')
plt.show()
```

## Visualizations - Part 3 (Differences)

For starters, it took fewer statements to create the same plots in Seaborn, especially when it came to subplots. In Matplotlib, each subplot had to be initialized with a plt.subplot statement, while in Seaborn the axis could just be passed to the plot statement as an argument.

Seaborn also spaced the plots out better, while Matplotlib jammed the plots up against each other, even overrunning the x-axis labels with the plot below it. That required me to manually set the plot spacing, while Seaborn seemed to do it more intelligently on its own.

The thing that Seaborn didn't do very well automatically was decide binwidth. In one of the plots, the binwidth was so small that at first it looked like an empty plot. Adjusting it manually fixed the problem. In a different plot, there was an outlier with a very large negative number; Matplotlib handled this with no problem, while Seaborn churned on it for a full five minutes before I had to kill the kernel and figure out what it was doing. Removing the outlier fixed the issue, but it was interesting that Matplotlib handled it while Seaborn was kind of like Tom Cruise in A Few Good Men: It couldn't handle the truth.


## Conclusions

Surprisingly, there seemed to be a positive correlation between the mean total grade and several variables: the amount of mouse movement in pixels, the number of keystrokes, the number of mouse clicks (both right and left), and the amount of time spent scrolling the mouse wheel. While it is doubtful that any of these factors directly influenced the student's grade, it is nonetheless interesting to note that students who performed better also demonstrated a greater amount of mouse activity.
